# Криптоалгоритмы

```python
# Шифр Атбаш
def atbash_encrypt_decrypt(text):
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    reverse_alphabet = alphabet[::-1]
    encrypted_text = ""
    
    for char in text.upper():
        if char in alphabet:
            index = alphabet.index(char)
            encrypted_text += reverse_alphabet[index]
        else:
            encrypted_text += char 
    return encrypted_text

# xtr упрощенный
def xtr_encrypt(text, key):
    encrypted_text = ""
    for char in text:
        encrypted_char = chr((ord(char) + key) % 256)
        encrypted_text += encrypted_char
    return encrypted_text

def xtr_decrypt(text, key):
    decrypted_text = ""
    for char in text:
        decrypted_char = chr((ord(char) - key) % 256)
        decrypted_text += decrypted_char
    return decrypted_text


def main():
    print("Выберите метод шифрования:")
    print("1. Симметричный (Атбаш)")
    print("2. Асимметричный (XTR)")
    choice = input("Введите номер метода (1 или 2): ")
    if choice == "1":
        text = input("Введите текст для шифрования/расшифровки: ")
        result = atbash_encrypt_decrypt(text)
        print("Результат (Атбаш):", result)
    elif choice == "2":
        text = input("Введите текст для шифрования: ")
        key = input("Введите ключ (число): ")
        try:
            key = int(key)
        except ValueError:
            print("Ошибка: ключ должен быть числом.")
            return
        encrypted_text = xtr_encrypt(text, key)
        print("Зашифрованный текст (XTR):", encrypted_text)
        decrypted_text = xtr_decrypt(encrypted_text, key)
        print("Расшифрованный текст (XTR):", decrypted_text)
    else:
        print("Ошибка: неверный выбор метода.")
```

### Что такое симметричные криптоалгоритмы?
Симметричные криптоалгоритмы используют один и тот же ключ как для шифрования, так и для расшифровки данных. Примеры: AES, DES, шифр Вернама.

### Что такое асимметричные криптоалгоритмы?
Асимметричные криптоалгоритмы используют пару ключей: открытый для шифрования и закрытый для расшифровки. Примеры: RSA, алгоритм Эль-Гамаля.

### Классификация симметричных криптоалгоритмов.
1.	По типу данных: блочные, поточные.
2.	По методу шифрования: подстановки, перестановки, комбинированные.
3.	По типу ключей: фиксированный, динамический.

### Классификация асимметричных криптоалгоритмов.
1.	По области применения: шифрование, цифровая подпись, обмен ключами.
2.	По математическим основам: алгоритмы на основе разложения на множители (RSA), алгоритмы на эллиптических кривых, алгоритмы на основе дискретного логарифма.

### Обобщенная схема симметричной криптосистемы.
1.	Сообщение шифруется с помощью симметричного ключа.
2.	Ключ передаётся между сторонами защищённым каналом.
3.	Получатель использует тот же ключ для расшифровки.

### Блочные и поточные шифры.
Блочные шифры обрабатывают данные блоками фиксированной длины (например, AES). Поточные шифры работают с данными побитно или посимвольно (например, RC4).

### Что понимается под шифрованием информации методом гаммирования?
Гаммирование — это метод шифрования, в котором к открытому тексту применяется побитовое или посимвольное сложение с ключевой последовательностью (гаммой).

### Что называется ключом в криптосистеме?
Ключ — это уникальная информация (набор данных), используемая для шифрования и расшифровки сообщений.

### Разовый блокнот.
Разовый блокнот — это шифр, использующий случайно сгенерированную гамму длиной, равной длине сообщения. Он абсолютно криптостойкий, если гамма используется только один раз.

### Что такое криптостойкость системы?
Криптостойкость — это мера защищённости криптосистемы от взлома. Она зависит от сложности алгоритма, длины ключа и устойчивости к атакам.

### Правила использования ключа.
1.	Ключ должен быть случайным.
2.	Должен передаваться по защищённому каналу.
3.	Не использовать один ключ для разных сообщений.
4.	Должен быть недоступен третьим лицам.
