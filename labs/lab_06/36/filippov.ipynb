# Задание 1

Реализуйте рекурсивный алгоритм, распечатывающий все перестановки натуральных чисел 1, 2, …, N. по одному разу. Перестановка - последовательность длины N, в которую каждое из этих чисел входит по одному разу.

```python
def generate_permutations(current, remaining):
    if not remaining:
        print(current)
        return
    for i in range(len(remaining)):
        generate_permutations(current + [remaining[i]], remaining[:i] + remaining[i + 1:])

def print_permutations(n):
    numbers = list(range(1, n + 1))
    generate_permutations([], numbers)

n = 6
print_permutations(n)
```

Реализовать алгоритм без использования рекурсии

```python
def generate_permutations(n):
    sequence = list(range(1, n + 1))
    while True:
        print(sequence)
        i = len(sequence) - 2
        while i >= 0 and sequence[i] >= sequence[i + 1]:
            i -= 1
        if i == -1:
            break
        j = len(sequence) - 1
        while sequence[j] <= sequence[i]:
            j -= 1
        sequence[i], sequence[j] = sequence[j], sequence[i]
        sequence = sequence[:i + 1] + list(reversed(sequence[i + 1:]))

n = 6
generate_permutations(n)
```

# Задание 3

Eсли sys.getrecursionlimit() = 1000, то максимальный стек - 1000 вызовов, а N при этом максимально равно 1000


```python
# реализация с промежуточным сохранением
def generate_permutations_cached(n):
    cache = {}
    def helper(current, remaining):
        key = (tuple(current), tuple(remaining))
        if key in cache:
            return cache[key]
        if not remaining:
            print(current)
            return
        for i in range(len(remaining)):
            helper(current + [remaining[i]], remaining[:i] + remaining[i + 1:])
        
        cache[key] = True 
    
    numbers = list(range(1, n + 1))
    helper([], numbers)
    
print(generate_permutations_cached(6))
```

```python
def memorize(func):
    cache = {}
    def wrapper(*args):
        hashable_args = tuple(tuple(arg) if isinstance(arg, list) else arg for arg in args)
        if hashable_args in cache:
            return cache[hashable_args]
        result = func(*args)
        cache[hashable_args] = result
        return result
    return wrapper

@memorize
def generate_permutations_recursive(current, remaining):
    if not remaining:
        print(current)
        return
    for i in range(len(remaining)):
        generate_permutations_recursive(current + [remaining[i]], remaining[:i] + remaining[i + 1:])

n = 6
generate_permutations_recursive([], list(range(1, n + 1)))
```

```python
# сравнение производительности алгоритмов
import time

def test_recursive(n):
    def generate(current, remaining):
        if not remaining:
            return
        for i in range(len(remaining)):
            generate(current + [remaining[i]], remaining[:i] + remaining[i + 1:])
    start = time.time()
    generate([], list(range(1, n + 1)))
    return time.time() - start

def test_iterative(n):
    def next_permutation(sequence):
        i = len(sequence) - 2
        while i >= 0 and sequence[i] >= sequence[i + 1]:
            i -= 1
        if i == -1:
            return False
        j = len(sequence) - 1
        while sequence[j] <= sequence[i]:
            j -= 1
        sequence[i], sequence[j] = sequence[j], sequence[i]
        sequence[i + 1:] = reversed(sequence[i + 1:])
        return True

    sequence = list(range(1, n + 1))
    start = time.time()
    while True:
        if not next_permutation(sequence):
            break
    return time.time() - start

n = 8
a = test_recursive(n)
b = test_iterative(n)
print("Рекурсивный алгоритм:", a, "секунд")
print("Итеративный алгоритм:", b, "секунд")
print("Выиграш в скорости по сравнению с рекурсивным алгоритмом на", a/b * 100 - 100, "процентов")
```
